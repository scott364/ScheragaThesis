```python
%load_ext autoreload
%autoreload 2
from context import rmlib
import rmlib.rmtools as rm
import pickle
import gzip
```

# Load Models

```python
# Set file name
file = 'collected_data/set_8' # Don't include file type

# Load setup image
img = rm.images.load_jpeg(file+'.jpeg')
print('Setup Image:')
rm.viewer.show_image(img)

f = gzip.open(file+'.pklz','rb')
models = pickle.load(f)
f.close()

print('Number of models loaded:',len(models))
```

# Extract images

```python
path = './collected_data/set_8_color/'
for i, model in enumerate(models):
    print(i)
    name = str(model.created_datetime)
    rm.images.save_jpeg(model.color_image, path+name+'.jpeg')
    
```
